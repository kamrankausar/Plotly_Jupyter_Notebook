```python
import plotly as py # import the plotly
import plotly.graph_objs as go # Import the graph object
from plotly.graph_objs  import Scatter, Layout # Import Graph Type and Layout of the Graph
py.offline.init_notebook_mode(connected = True) # For using Plotly as offline
```

```python
# Create a trace
trace1 = go.Scatter(
    x = [1, 2, 3, 4],
    y = [4, 3, 2, 1]) 
```

```python
data1 = [trace1]
```

```python
layout1 = go.Layout(
    title='<b>SIMPLE EXAMPLE</b>',
    yaxis=dict(
        title='<i>(Y-Axis)</i>'
    ),
    xaxis=dict(
        title='<i>(X-Axis)</i>'
    )
)
```

```python
fig1 = go.Figure(data = data1, layout = layout1)
py.offline.iplot(fig1)
```

```python
import plotly.plotly as py
import plotly.graph_objs as go
import plotly as py
import plotly.graph_objs as go
py.offline.init_notebook_mode(connected = True)
```

```python
import numpy as np
```

```python
N = 500
random_x = np.linspace(0,1,N)
random_y = np.random.randn(N)
```

```python
# Create a trace
trace = go.Scatter(
    x = random_x,
    y = random_y)
```

```python
data = [trace]
```

```python
layout = go.Layout(
    title='<b>SIMPLE EXAMPLE</b>',
    yaxis=dict(
        title='<i>(Volts)</i>'
    ),
    xaxis=dict(
        title='<i>(nanoseconds)</i>'
    )
)
```

```python
fig = go.Figure(data = data, layout = layout)
py.offline.iplot(fig)
```
